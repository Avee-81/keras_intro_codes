```python
from keras.models import Model
from keras.utils import to_categorical
from keras.layers import Input,Dense,concatenate


import numpy as np
```

Initialize the Hyper-params


```python
no_of_epochs = 100
```

```python
inputs= Input(shape=(64,))

shared_layer = Dense(64,activation="relu")

x = shared_layer(inputs)
x = shared_layer(x)

preds = Dense(10,activation="softmax")(x)
```

Initialize the Model

```python
model = Model(inputs =inputs, outputs=preds)

model.summary()
```

Generate Dummy Data

```python
x_train = np.random.random((1000,64))
y_cat = np.random.randint(10,size = (1000,1))
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, to_categorical(y_cat), epochs=no_of_epochs, batch_size=32 )
```

```python
model.save('shared_layer.h5')
```
